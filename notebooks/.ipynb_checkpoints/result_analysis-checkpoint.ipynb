```python
# Lire les données des AWS S3 bucket
import os
from pyspark.sql import SparkSession

# Créer une session Spark avec les configurations AWS
spark = SparkSession \
    .builder \
    .appName("ReadS3Data") \
    .config("spark.jars", "/usr/local/spark/jars/hadoop-aws-3.3.1.jar,/usr/local/spark/jars/aws-java-sdk-bundle-1.11.1026.jar") \
    .config("spark.hadoop.fs.s3a.access.key", os.environ['AWS_ACCESS_KEY_ID']) \
    .config("spark.hadoop.fs.s3a.secret.key", os.environ['AWS_SECRET_ACCESS_KEY']) \
    .config("spark.hadoop.fs.s3a.endpoint", "s3.amazonaws.com") \
    .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem") \
    .config("spark.hadoop.fs.s3a.aws.credentials.provider", "org.apache.hadoop.fs.s3a.SimpleAWSCredentialsProvider") \
    .getOrCreate()
```

```python
# Lire les données
df = spark.read.parquet(os.environ['AWS_S3_RESULT_FOLDER'])

# Afficher les données
df.show()
```

```python
# Analyses par support_team
df.groupBy("support_team").sum("count").show()

# Visualisation avec matplotlib
import matplotlib.pyplot as plt

# Convertir en pandas pour la visualisation
pdf = df.toPandas()

# Créer un graphique
plt.figure(figsize=(10, 6))
pdf.groupby('support_team')['count'].sum().plot(kind='bar')
plt.title('Tickets par équipe de support')
plt.xlabel('Équipe')
plt.ylabel('Nombre de tickets')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
